```python
namelist1=[]
typelist1=[]
for i in range(0,24):
    response4=requests.get('https://potguide.com/strain-profiles/?page={}'.format(i))
    soup4=BeautifulSoup(response4.content,'html.parser')
    names=soup4.findAll('span',{'itemprop':'name'})
    for n in names:
        namelist1.append(n.text)
    types=soup4.findAll('small',{'itemprop':'description'})
    for t in types:
        typelist1.append(t.text)
moretypes={'name':namelist1,'type':typelist1}
```

```python
typedf2=pd.DataFrame(moretypes)
```

```python
alltypesdf=pd.concat([alltypesdf,typedf2])
```

```python
namelist2=[]
typelist2=[]
thclist2=[]
for i in range(0,62):
    response5=requests.get('https://www.wikileaf.com/strains/?page={}'.format(i))
    soup5=BeautifulSoup(response5.content,'html.parser')
    names=soup5.findAll('h5',{'class':'name disp-title'})
    for n in names:
        namelist2.append(n.text)
    types=soup5.findAll('p',{'class':'tag'})
    for t in types:
        typelist2.append(t.text)
    thcs=soup5.findAll('p',{'class':'desc'})
    for t in thcs:
        thclist2.append(t.text[4:6])
moretypes1={'name':namelist2,'type':typelist2}

typedf3=pd.DataFrame(moretypes1)

alltypesdf=pd.concat([alltypesdf,typedf3])
```
