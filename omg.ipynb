```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
h2o,thc,cbd,pine,myrc,lina,care,euca,cary=[],[],[],[],[],[],[],[],[]
namelist=[]
infolist=[]
for i in range(1,44):
    response=requests.get('http://ocpdb.pythonanywhere.com/ocpdb/?page={}'.format(i))
    soup=BeautifulSoup(response.content,'html.parser')
    names=(soup.findAll('p',{'class':'h5'}))
    for n in names:
        namelist.append(n.text.strip('\n\t\t\t\t\t\t\t\t\t\t\t\t\t').lower())
    info=soup.findAll('td',{'class':'record'},'strong')
    for i in info:
        infolist.append(i.text.strip('\n').strip('\n\t\t\t\t\t\t\t\t\t\t\t'))
for x in infolist:
    if x.startswith('H2O'):
        h2o.append(x[-5:-1])
    if x.startswith('Total THC'):
        y=x[11:16]
        thc.append(y.strip('\\xa0'))
    if x.startswith('Total CBD'):
        cbd.append(x[11:16])
    if x.startswith('Î±-Pinene'):
        pine.append(x[10:14])
    if x.startswith('Myrcene'):
        myrc.append(x[9:13])
    if x.startswith('Linalool'):
        lina.append(x[10:15])
```

```python
namelist=namelist[:-3]
print(len(namelist),len(thc),len(cbd),len(pine),len(myrc),len(lina),len(h2o))
```

```python
dic={'name':namelist,'thc':thc,'cbd':cbd,'pinene':pine,'myrcene':myrc,'linalool':lina}
```

```python
df=pd.DataFrame(dic)
```
