```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response2=requests.get('https://www.potbotics.com/learn/strains')
soup2=BeautifulSoup(response2.content,'html.parser')
#find all names
names=soup2.findAll('h4',{'class':'card-title'})
namelist=[]
for n in names:
    namelist.append(n.text.lower())
types=soup2.findAll('h5')
typelist=[]
for t in types:
    typelist.append(t.text.lower())
typelist=typelist[:-3]
thcs=soup2.findAll('p',{'class':'amount thc'})
thclist=[]
for t in thcs:
    thclist.append(t.text)
cbds=soup2.findAll('p',{'class':'amount cbd'})
cbdlist=[]
for c in cbds:
    cbdlist.append(c.text)
cbns=soup2.findAll('p',{'class':'amount cbn'})
cbnlist=[]
for c in cbns:
    cbnlist.append(c.text)
cbcs=soup2.findAll('p',{'class':'amount cbc'})
cbclist=[]
for c in cbcs:
    cbclist.append(c.text)
thcvs=soup2.findAll('p',{'class':'amount thcv'})
thcvlist=[]
for c in thcvs:
    thcvlist.append(c.text)
cbgs=soup2.findAll('p',{'class':'amount cbg'})
cbglist=[]
for c in cbgs:
    cbglist.append(c.text)
moreweed={'name':namelist,'type':typelist,'thc':thclist,'cbd':cbdlist,'cbn':cbnlist,'cbc':cbclist,'thcv':thcvlist,'cbg':cbglist}
df2=pd.DataFrame(moreweed)
df2
```

```python
namelist=[]
typelist=[]
thclist=[]
for i in range(1,62):
    response=requests.get('https://www.wikileaf.com/strains/?page={}'.format(i))
    soup=BeautifulSoup(response.content,'html.parser')
    names=soup.findAll('h5',{'class':'name disp-title'})
    types=soup.findAll('p',{'class':'tag'})
    thcs=soup.findAll('p',{'class':'desc'})
    for n,t,x in zip(names,types,thcs):
        namelist.append(n.text.lower())
        typelist.append(t.text.lower())
        thclist.append(x.text[3:6])
```

```python
moreinfo={'name':namelist,'type':typelist,'thc':thclist}
df3=pd.DataFrame(moreinfo)
# df3[df3['name']=='green crack']['type']
```

```python
h2o,thc,cbd,pine,myrc,lina,care,euca,cary=[],[],[],[],[],[],[],[],[]
namelist=[]
infolist=[]
for i in range(1,44):
    response=requests.get('http://ocpdb.pythonanywhere.com/ocpdb/?page={}'.format(i))
    soup=BeautifulSoup(response.content,'html.parser')
    names=(soup.findAll('p',{'class':'h5'}))
    for n in names:
        namelist.append(n.text.strip('\n\t\t\t\t\t\t\t\t\t\t\t\t\t').lower())
    info=soup.findAll('td',{'class':'record'},'strong')
    for i in info:
        infolist.append(i.text.strip('\n').strip('\n\t\t\t\t\t\t\t\t\t\t\t'))
for x in infolist:
    if x.startswith('H2O'):
        h2o.append(x[-5:-1])
    if x.startswith('Total THC'):
        y=x[11:16]
        thc.append(y.strip('\\xa0'))
    if x.startswith('Total CBD'):
        cbd.append(x[11:16])
    if x.startswith('α-Pinene'):
        pine.append(x[10:14])
    if x.startswith('Myrcene'):
        myrc.append(x[9:13])
    if x.startswith('Linalool'):
        lina.append(x[10:15])
```

```python
namelist=namelist[:-3]
dic={'name':namelist,'thc':thc,'cbd':cbd,'pinene':pine,'myrcene':myrc,'linalool':lina}
```

```python
df4=pd.DataFrame(dic)
```
